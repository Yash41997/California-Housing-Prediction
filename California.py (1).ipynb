```python
housing_num=pd.get_dummies(data,columns=["ocean_proximity"],dtype="float64")
housing_num.interpolate(inplace=True)
housing_num.info()
```

```python
import seaborn as sns
sns.heatmap(housing_num.isnull(),cbar=False,yticklabels=False,cmap="Set3")
```

```python
housing_num.info()
y=housing_num["median_house_value"]
x=housing_num.drop(columns="median_house_value")
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=101)
train_x
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(train_x,train_y)
```

```python
predicted_y=lin.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(predicted_y,test_y)
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec=DecisionTreeRegressor()
dec.fit(train_x,train_y)
pred_y_DecTree=dec.predict(test_x)
```

```python

error1=mean_absolute_error(pred_y_DecTree,test_y)
error1
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(max_features=6,n_estimators=500)
forest_reg.fit(train_x,train_y)
```

```python
pred_y_RandFor=forest_reg.predict(test_x)
error2=mean_absolute_error(pred_y_RandFor,test_y)
error2
```
